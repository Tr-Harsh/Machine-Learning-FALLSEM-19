```python
plt.figure(figsize=(24,24))
for index,column in enumerate(data.columns):
    if(column == "grade"):
        continue
    plt.subplot(6,6,index+1)
    sb.violinplot(x='grade',y=column,data=data)
```

```python
plt.figure(figsize=(24,24))
for index,column in enumerate(data.columns):
    if(column == "grade"):
        continue
    plt.subplot(6,6,index+1)
    sb.boxplot(x='grade',y=column,data=data)
```
